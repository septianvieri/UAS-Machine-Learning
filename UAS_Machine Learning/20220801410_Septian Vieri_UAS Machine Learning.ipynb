2.Jelaskan apa yang dimaksud dengan unsupervised learning

Unsupervised learning adalah jenis pembelajaran mesin di mana model dilatih menggunakan data yang tidak berlabel. Ini berarti bahwa tidak ada output atau target yang diberikan untuk membantu model belajar. Sebaliknya, model mencoba menemukan pola atau struktur yang mendasari dalam data itu sendiri. Berikut adalah penjelasan lebih detail tentang unsupervised learning:

Karakteristik Unsupervised Learning

* Tidak Ada Label: Data yang digunakan dalam unsupervised learning tidak memiliki label atau target output. Model harus mengidentifikasi pola dalam data tanpa panduan eksplisit.
* Tujuan Utama: Mencari pola, kelompok, atau struktur dalam data untuk mendapatkan wawasan lebih lanjut atau mereduksi dimensionalitas data.
* Aplikasi: Clustering (pengelompokan), asosiasi, reduksi dimensionalitas.

Jenis-jenis Unsupervised Learning

A.Clustering (Pengelompokan):

* K-Means Clustering: Mengelompokkan data ke dalam k kelompok berdasarkan jarak terdekat ke pusat cluster.
* Hierarchical Clustering: Membangun hierarki kelompok dengan menggabungkan atau membagi cluster.
* DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Mengelompokkan titik data berdasarkan kepadatan titik di sekitarnya.

B.Association (Asosiasi):

* Apriori Algorithm: Mengidentifikasi aturan asosiasi dalam data, seperti menemukan produk yang sering dibeli bersama dalam transaksi ritel.
* Eclat Algorithm: Alternatif dari Apriori, menggunakan metode intersection untuk menemukan itemset yang sering muncul.

C.Dimensionality Reduction (Reduksi Dimensionalitas):

* Principal Component Analysis (PCA): Mengurangi jumlah dimensi dalam data dengan mengubahnya menjadi komponen utama yang memaksimalkan varians.
* t-Distributed Stochastic Neighbor Embedding (t-SNE): Digunakan untuk visualisasi data dengan mereduksi dimensi tinggi menjadi dua atau tiga dimensi.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
import seaborn as sns

# Download required NLTK data
nltk.download('vader_lexicon')
nltk.download('stopwords')
nltk.download('punkt')

# Define the text
text = """
Once upon a time in a faraway palace, there lived a beautiful princess
...
Finally they lived happily ever after.
"""

# Create a WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Tokenize the text
words = nltk.word_tokenize(text.lower())

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalnum() and word not in stop_words]

# Get word frequencies
word_freq = Counter(filtered_words)
common_words = word_freq.most_common(10)

# Create a DataFrame for visualization
df = pd.DataFrame(common_words, columns=['Word', 'Frequency'])

# Create a bar plot for word frequencies
plt.figure(figsize=(10, 5))
sns.barplot(x='Word', y='Frequency', data=df)
plt.title('Top 10 Most Common Words')
plt.show()
```

```python
# Sentiment analysis
sia = SentimentIntensityAnalyzer()
sentiment_scores = sia.polarity_scores(text)

# Display sentiment scores
sentiment_df = pd.DataFrame(list(sentiment_scores.items()), columns=['Sentiment', 'Score'])
plt.figure(figsize=(10, 5))
sns.barplot(x='Sentiment', y='Score', data=sentiment_df)
plt.title('Sentiment Analysis')
plt.show()
```
